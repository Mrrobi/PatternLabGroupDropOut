```
from sklearn.model_selection import KFold
```

```

pos_5289 = "/content/drive/My Drive/Pattern_Lab/npy/A.thaliana5289_pos_PseKNC.npy"
neg_5289 = "/content/drive/My Drive/Pattern_Lab/npy/A.thaliana5289_neg_PseKNC.npy"
pos_1000 = "/content/drive/My Drive/Pattern_Lab/npy/A.thaliana1000indep_pos_pseknc.npy"
neg_1000 = "/content/drive/My Drive/Pattern_Lab/npy/A.thaliana1000indep_neg_pseknc.npy"

x_train_pos = np.load(pos_5289) 
x_train_neg = np.load(neg_5289) 
x_test_pos = np.load(pos_1000) 
x_test_neg = np.load(neg_1000) 

y_train_pos = np.tile(1, 5289) 
y_train_neg = np.tile(0, 5289) 
y_test_pos = np.tile(1, 1000) 
y_test_neg = np.tile(0, 1000) 
```

```
seed = 40
np.random.seed(seed)
```

```
x_training = np.concatenate((x_train_pos, x_train_neg), axis= 0)
y_training = np.concatenate((y_train_pos, y_train_neg))
```

```
x_test = np.concatenate((x_test_pos, x_test_neg))
y_test = np.concatenate((y_test_pos, y_test_neg))
```

```
kf = KFold(n_splits=5, random_state=seed)
```

```

clf = RandomForestClassifier(n_estimators=250, random_state = seed)
val_accuracy_list = []
train_accuracy_list = []
sum_val=0
sum_train=0
i = 1
for train_index, val_index in kf.split(x_training):
  print("Fold : ",i)
  X_train, X_val = x_training[train_index], x_training[val_index]
  y_train, y_val = y_training[train_index], y_training[val_index]
  clf.fit(X_train, y_train)
  y_val_pred = clf.predict(X_val)
  val_accuracy = accuracy_score(y_val, y_val_pred)
  sum_val += val_accuracy
  val_accuracy_list.append(val_accuracy)
  y_train_pred = clf.predict(X_train)
  train_accuracy = accuracy_score(y_train, y_train_pred)
  sum_train += train_accuracy
  train_accuracy_list.append(train_accuracy)
  i += 1 
ave_val=sum_val/5
ave_train=sum_train/5
print("done")
for i in range(0, len(train_accuracy_list)):
  print("train_accuracy : ", train_accuracy_list[i], "\tVal accuracy : ", val_accuracy_list[i])
print("train_accuracy_ave : ",ave_train,"\tval accuracy_ave:",ave_val)
  
  
  
```
